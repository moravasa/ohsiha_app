```python
# Viikko 1 demo: harjoitellaan lyhyesti datan pyörittelyä ('wrangling') 
# Pandas-kirjaston avulla sekä yksinkertaista visualisointia matplotlibillä.

import pandas as pd
import matplotlib.pyplot as plt

# hyödynnetään kurssin ensimmäisessä demossa käytettyä datasettiä:
url = 'https://github.com/jodatut/2020/raw/master/koodiesimerkit/SalesJan2009.csv'

# luetaan csv-tiedosto dataframeen
df = pd.read_csv(url)
# luodaan uusi dataframe, johon poimitaan vain tietyt sarakkeet
df_transaktiot = df[['Transaction_date', 'Product', 'Price', 'Country']].copy()

# muunnetaan transaction_date datetimeksi
df_transaktiot['Transaction_date'] = pd.to_datetime(df['Transaction_date'])
# oletetaan että hinnat ovat dollareina ja muunnetaan ne euroiksi
df_transaktiot['Price'] = df_transaktiot['Price'] * 0.85
# nimetään sarakkeet uudelleen
df_transaktiot.rename(columns={'Transaction_date': 'aikaleima', 'Product':'tuote', 'Price':'hinta_eur', 'Country':'maa'}, inplace=True)
# tulostetaan 5 ensimmäistä riviä
print(df_transaktiot.head())
```

```python
# lasketaan tammikuun 2009 kokonaismyynti euroina Saksassa
df_transaktiot_saksa = df_transaktiot[df_transaktiot['maa'] == 'Germany']
df_myynti_saksa_sum = sum(df_transaktiot_saksa['hinta_eur'])
print("Saksan myynti tammikuussa 2009 oli " + str(df_myynti_saksa_sum) + " eur.")
```

```python
# haetaan datasta päivittäinen myynti euroina USA:ssa
df_transaktiot_USA = df_transaktiot[df_transaktiot['maa'] == 'United States'].copy()

anna_paiva = lambda x: x.day
df_transaktiot_USA['paiva'] = df_transaktiot_USA['aikaleima'].apply(anna_paiva)
df_paivamyynti_USA=df_transaktiot_USA.groupby('paiva')['hinta_eur'].sum()

# piirretään kuvaaja USA:n päivittäisestä myynnistä euroina tammikuussa 2009

#print(paivamyynti_USA)
df_paivamyynti_USA.plot.bar()
plt.ylabel('Myynti eur')
plt.xlabel('Päivä')
plt.title('Tammikuun 2009 myynti USA:ssa')
plt.show()
```

*Demo viikon 2 aikana oppimistani asioista*

Tämän viikon demossa kertaan datan poimimista API-rajapinnasta. Aihepiiri on jo jossain määrin tuttu Ohjelmallinen sisällönhallinta -kurssilta (v. 2019), mutta kertauksen vuoksi testailen tässä jotain itselleni uutta rajapintaa. Tässä demossa hyödynnän varsin ajankohtaista API:a, eli THL:n Sampo-rajapintaa, jonka kautta saadaan poimittua varmistettujen koronatapausten määrät Suomessa.



```python
# Viikko 2 demo: lähtödatan poiminta avoimen API-rajapinnan kautta
# Haetaan THL:n rajapinnasta kuluvan viikon koronatartunnat HUS-alueella

# THL avoin rajapinta: varmistetut koronatapaukset Suomessa:
# https://thl.fi/fi/tilastot-ja-data/aineistot-ja-palvelut/avoin-data/varmistetut-koronatapaukset-suomessa-covid-19-

# Apuna tässä käytetty myös sivustoa https://realpython.com/python-api/

import requests
import datetime

# haetaan nykyisen viikon numero
current_date = datetime.date.today()
year, week, day = current_date.isocalendar()

# koronatartunnat Helsingin ja Uudenmaan sairaanhoitopiirin (HUS) alueella:
api_addr = 'https://sampo.thl.fi/pivot/prod/fi/epirapo/covid19case/fact_epirapo_covid19case.json?row=hcdmunicipality2020-445193.&column=dateweek20200101-509030&filter=measure-444833'

print("Haetaan HUS-alueella ilmoitetut koronatapaukset viikolla " + str(week) + " / " + str(year) + " API-rajapinnasta...")

try: 
  # yritetään hakea dataa api-rajapinnasta
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
  response = requests.get(api_addr, headers=headers)
  
except:
  print("Tietojen haku epäonnistui!")

if response.status_code != 200:
  print("Tietojen haku epäonnistui (HTTP " + str(response.status_code) + ")" )
else:
  # haku onnistui, http 200
  json_response = response.json()
  codes=json_response['dataset']['dimension']['dateweek20200101']['category']['label']
  print("Tietojen poiminta onnistui.")

  # haetaan aineistosta kuluvaa viikkoa vastaava koodi 
  search_value = "Vuosi " + str(year) + " Viikko " + str(week)
  
  code = 0
  index = 0

  for key in codes:
    if codes[key] == search_value:
      code = key
      break
  
  # haetaan viikkokoodia vastaava id
  # id:n perusteella saadaan tartuntamäärät ('value')

  indices=json_response['dataset']['dimension']['dateweek20200101']['category']['index']
  for key in indices:
    if key == code:
      index = indices[key]
      break
  
  if index == 0:
    print("Virhe: aineistossa ei ole dataa halutulle viikolle!")
  else:
    print ("*** debug: viikon koodi " + str(code) + ", indeksi: " + str(index))
    # haetaan ID:tä vastaava tartuntamäärälukema
    covid_cases=json_response['dataset']['value']
    index_str = str(index)
    print("Koronatartunnat viikolla " + str(week) + ": " + str(covid_cases[str(index)]))
```

Koneoppimisalgoritmit voidaan jakaa neljään eri tyyppiin: ohjattu, ohjaamaton, puoliohjattu sekä vahvistettu oppiminen. Ohjatussa oppimisessa kone pyrkii ennustamaan tiettyä luokkaa (label) sille annetun oikeat vastaukset sisältävän opetusaineiston perusteella. Sen sijaan ohjaamattomassa oppimisessa koneelle ei anneta oikeita vastauksia, vaan se pyrkii löytämään datasta jonkinlaista säännönmukaisuutta, kuten esimerkiksi samankaltaisten tapausten ryppäitä (clusters). Vahvistusoppimisessa kone saa palautetta siitä, kuinka onnistuneesti se toimii. Kone oppii tunnistamaan yhteyksiä ympäristön tilan, oman toimintansa ja saadun palautteen välillä ja näin tekemään parempia valintoja. Puoliohjattu koneoppiminen sijoittuu ohjatun ja ohjaamattoman koneoppimisen välimaastoon. Siinä mallia opetetaan sekä luokitetulla että luokittelemattomalla aineistolla.

<br>

Yksi tärkeä koneoppimiseen liittyvä käsite ovat <i>piirteet</i> (features), joihin laskennallinen analytiikka perustuu. Piirteet tarkoittavat samaa asiaa kuin syötteen (input) muuttujat. Ne voivat olla binäärisiä, kategorisia tai jatkuvia. Esimerkiksi asunnon hinnan määritteleviä piirteitä ovat mm. asunnon sijainti, pinta-ala, kunto ja rakennusvuosi. 


<i>Piirteiden jalostaminen</i> (feature engineering) on prosessi, jossa ihmisten asiantuntemusta hyödyntämällä lähtödatasta irrotetaan ja jalostetaan koneoppimisen kannalta olennaiset piirteet. Piirteiden jalostaminen on tärkeä vaihe koneoppimisen työnkulussa, sillä se vaikuttaa oleellisesti koneoppimisalgoritmien tehokkuuteen ja ennustetarkkuuteen. Piirteiden jalostamisessa voidaan esimerkiksi laskea uusia piirteitä olemassa olevista piirteistä tai pyrkiä vähentämään piirteiden määrää mm. pääkomponenttianalyysin avulla. Jalostettuja piirteitä voidaan muodostaa myös luonnollista kieltä sisältävälle tekstiaineistolle. Tunnesävyanalyysi on yksi esimerkki analytiikan keinoin tuotetusta jalostetusta piirteestä.

<br>

Tällä luentoviikolla fokus oli <i>ohjatussa oppimisessa</i>, jossa pyritään ennustamaan tiettyä muuttujaa ("label"). Ennustettava muuttuja voi olla esimerkiksi luokka (marja/vihannes/hedelmä), binäärinen luokka (ehjä/rikki) tai numeerinen arvo, kuten Airbnb-asunnon hinta (82 euroa). Ohjattu oppinen vaatii aina opetusdataa eli joukkoa esimerkkejä, joista jokaisen kohdalla on kerrottu oikea vastaus. Lyhykäisyydessään ohjatun oppimisen työnkulku etenee seuraavasti: 
1. "Raakadata" siivotaan ja jalostetaan piirteet -> piirrematriisi
2. Koneoppimismalli opetetaan piirrematriisin sekä piirteitä vastaavien labeleiden eli oikeiden vastausten avulla
3. Opetetun mallin suorituskykyä arvioidaan testidatan avulla
4. Viedään malli tuotantoon. Uudesta datasta irrotetaan piirrevektori, joka syötetään ennustemalliin. Ennustemallin ulostulona saadaan ennustettavan muuttujan arvo ("label").

Yksi liiketoiminnassa hyödynnetty ohjatun koneoppimisen sovelluskohde on ns. asiakaspoistuma-analyysi, jossa ideana on selvittää ne asiakkaat, jotka ovat vaarassa poistua sekä poistuman todennäköisyys ja syy. Poistuma-analyysi soveltuu hyvin esimerkiksi sähkönmyyntiyhtiöille, teleoperaattoreille, kuntosaleille yms. sopimussuhteeseen perustuvaan kuluttajaliiketoimintaan. Koneoppimismallin opetusdataksi tarvitaan historiatietoa poistuneista ja ei-poistuneista asiakkaista (="label"). Opetusdatan piirteinä käytetään sekä asiakkaiden taustatietoja (esim. ikä, sukupuoli, asiakkuuden kesto) että heidän käyttäytymiseensä liittyviä tietoja (esim. asiakaspalautteet ja yhteydenotot asiakaspalveluun). Muodostetun koneoppimismallin avulla pystytään tunnistamaan todennäköisesti lähiaikoina poistumassa olevat asiakkaat sekä poistumiseen vaikuttaneita syitä. Poistumauhan alla olevat asiakkaat kontaktoidaan ja pyritään säilyttämään heidät yrityksen asiakkaina. Lisäksi asiakaspoistuman syyt tunnistamalla voidaan pyrkiä vähentämään asiakaspoistumaa esimerkiksi kehittämällä tuotteita ja asiakaspalvelun laatua. Asiakastyytyväisyyden paranemisen lisäksi saavutetaan taloudellista hyötyä, sillä tyypillisesti uuden asiakkaan hankkiminen on huomattavasti kalliimpaa kuin vanhan asiakkaan pitäminen.

```python
# Demo viikko 3
# Yhden muuttujan lineaarinen regressio yksinkertaisena koneoppimisen mallina

# Demon aineistona on oluen kulutus Saõ Paulon yliopiston alueella vuonna 2015
# https://www.kaggle.com/dongeorge/beer-consumption-sao-paulo
# Sama aineisto saadaan ladattua (helpommin) Githubista:
# (https://raw.githubusercontent.com/fiapteamrocket/python-para-ia/master/consumo_cerveja.csv)

# Pyritään ennustamaan Saõ Paulon yliopiston alueen oluen kulutusta ulkoilman
# lämpötilan perusteella

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import pandas as pd

data_path = 'https://raw.githubusercontent.com/fiapteamrocket/python-para-ia/master/consumo_cerveja.csv'

# luetaan aineisto pandas dataframeen
df_beer = pd.read_csv(data_path, delimiter=';')
df_beer.rename(columns={'Data':'Date', 'Temperatura Media (C)': 'Avg_temp', 'Final de Semana': 'Weekend', 'Consumo de cerveja (litros)': 'Beer_consumption_l'}, inplace=True)
X = df_beer[['Avg_temp']]
y = df_beer['Beer_consumption_l']

# jaetaan aineisto opetus- ja validointiaineistoon
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# piirretään alkup. aineiston hajontakaavio (scatter diagram) 
# ulkolämpötila vs. oluen kulutus
plt.scatter(X['Avg_temp'],y)

# yhden muuttujan lineaarinen regressio, selittävänä muuttujana on keskilämpötila
# selitettävä muuttuja on oluen kulutus litroina
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# piirretään regressiosuora
y_pred=lin_reg.predict(X_train)
plt.plot(X_train, y_pred, color='red')
plt.show()

# testataan mallia testiaineistolla
predicted=lin_reg.predict(X_test)
print("Ennusteen keskivirhe: ", metrics.mean_absolute_error(y_test, predicted))

out_temp=input("Anna ulkoilman lämpötila (C): ")

predicted = lin_reg.predict([[float(out_temp)]])
print("Ennustettu oluen kulutus (litraa): " + str(round(predicted[0],1)))
```

```python
# Demo viikko 4: AirBnb-datan sisäänluku ja yksinkertainen hinnan ennustin
#
# Harjoitustyössä tulen käyttämään ainakin yhtenä datalähteenä Inside Airbnb-dataa
# joten tällä viikolla demoillaan ko. palvelusta saatavan datan sisäänlukua.
# Tässä demossa käytän Kööpenhaminan Airbnb-dataa (h-työssä käytän Wienin dataa) 
# Aineistona tässä "listings.csv - summary information and metrics"
#
# Lisäksi aineistoon lasketaan uusi piirre "etäisyys keskustasta" sekä
# testaillaan asunnon hinnan ennustamista lineaarisen 
# monimuuttujaregression avulla.
# 
# Demon laatimisessa on hyödynnetty opintojakson materiaalista löytyvää esimerkkianalyysia:
# https://github.com/InfoTUNI/joda2021/blob/master/koodiesimerkit/airnbn/python_scikit_airbnb.ipynb


import pandas as pd
import haversine as hs
from matplotlib import pyplot as plt
from math import radians, cos, sin, asin, sqrt
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split
import sklearn.linear_model as linear_model

# luetaan datatiedosto inside airbnb:stä
# Kööpenhaminan Airbnb data, joka on kerätty 10.4.2021 
DATAFILE = 'http://data.insideairbnb.com/denmark/hovedstaden/copenhagen/2021-01-30/visualisations/listings.csv'

# määritetään pandas dataframeen luettavat sarakkeet
columns = ['price', 'latitude', 'longitude', 'room_type', 'neighbourhood', 'number_of_reviews']
df = pd.read_csv(DATAFILE, usecols=columns)

# valitaan vain kohteet, joissa huonetyyppi on 'Entire home/apt'
df = df[df.room_type == 'Entire home/apt']

# tulostetaan perustietoja hinta-sarakkeesta
df['price'].describe()

```

```python
# suodatetaan aineistosta 'outlierit' eli kohteet, joissa hinta hinta > 5000 DKK
locations = df[(df.price > 0) & (df.price <= 5000)]

locations['price'].describe()
# dataframeen jää 7204 kohdetta; huoneistojen keskihinta on 1054 DKK eli noin 140 euroa
```

```python
# Lasketaan aineistoon uusi piirre, "etäisyys keskustasta"
# Kööpenhaminan keskustan koordinaatit ovat 55.676098, 12.568337
# (lähde: https://www.latlong.net/place/copenhagen-denmark-4775.html)

city_centre_lat = 55.676098
city_centre_lon = 12.568337

# Suodatetaan aineistosta ensin pois kohteet, joista puuttuu jompi kumpi 
# koordinaattitieto tai se on selvästi virheellinen
locations = locations[(locations.latitude >= 55) & (locations.latitude < 56) & \
                      (locations.longitude >= 12) & (locations.longitude < 13)]

# lasketaan jokaiselle kohteelle etäisyys keskustasta kilometreinä
locations['city_centre_dist'] = df.apply(lambda row: \
                                         hs.haversine((city_centre_lon, city_centre_lat), (row.longitude, row.latitude)), axis=1)

# muunnetaan kaupunginosatieto numeeriseen muotoon ja pudotetaan pois tietyt
# sarakkeet
n_dummies = pd.get_dummies(locations.neighbourhood)
bnb_data = pd.concat((locations.drop(['neighbourhood', 'room_type', 'latitude', \
                                      'longitude'], axis=1), \
                     n_dummies.astype(int)), axis=1)
```

```python
# kokeillaan hinnan ennustamista jäljelle jääneiden piirteiden avulla
# (arvosteluiden lkm, etäisyys keskustasta, kaupunginosa)

X_train, X_test, y_train, y_test = train_test_split(bnb_data.drop(['price'], axis=1), \
                                                    bnb_data.price, test_size=0.2, random_state=20)

# käytetään koneoppimisen mallina Bayesin lineaarista regressiota
model = linear_model.BayesianRidge().fit(X_train, y_train)
err = metrics.median_absolute_error(y_test, model.predict(X_test))
print("median absolute error: " + str(err))

# tarkastellaan lopuksi mallin piirteiden suhteellista tärkeyttä
importance = abs(model.coef_)
importance = 100.0 * (importance / importance.max())
list_importance = list(importance)
list_features = list(X_train.columns)
data_tuples = list(zip(list_features, list_importance))
feature_importance = pd.DataFrame(data_tuples, columns=['Feature', 'importance'])
feature_importance=feature_importance.sort_values(by=['importance'])
plt.barh(feature_importance['Feature'], feature_importance['importance'])
```

Tuloksista nähdään, että absoluuttisen virheen mediaani on noin 280 DKK eli kovin hyvään ennustetarkkuuteen näiden piirteiden avulla ei päästy.

Lisäksi voidaan havaita, että eniten hintaan vaikuttava tekijä on sijanti kaupunginosassa "Indre By" eli ydinkeskusta. Sen sijaan absoluuttinen linnuntietä mitattu etäisyys keskustaan (city_centre_dist) vaikuttaa asunnon hintaan melko vähän.

```python
# Luonnollisen kielen analyysin sovelluksena testailen tunnesävyanalyysia.
# Tässä demossa hyödynnetään Pythonin Flair-kirjastoa, jossa on valmiina IMDB-
# datasetin pohjalta koulutettu malli tunnesävyanalyysiin (-> ei välttämättä
# toimi kovin hyvin muunlaisen tekstiaineiston kanssa..)
# https://towardsdatascience.com/text-classification-with-state-of-the-art-nlp-library-flair-b541d7add21f

# Asennetaan ensin tarvittaessa Flair-kirjasto
!pip install flair
```

```python
import flair
import pandas as pd

# Otetaan käyttöön Flairissa valmiina oleva englanninkielinen luokittelija
sentiment_model = flair.models.TextClassifier.load('en-sentiment')
```

```python
# Testataan luokittelijan toimintaa ao. esimerkkilauseella:
example_sentence = " I was not even able to watch one episode, it's super cheesy, wooden acting, the" + \
                   " dialogues lameness was unpreditable, not because they are original but because it's so deeply cliche that you wouldn't expect it."

# 'tokenisoidaan' lause ja talletetaan se Flairin sisäiseen formaattiin
sentence = flair.data.Sentence(example_sentence, use_tokenizer=True)

# Tunnesävyanalyysissa ei kannattane poistaa hukkasanoja (stopwords),
# sillä lauseen merkitys voi muuttua.
print(sentence)
sentiment_model.predict(sentence)

print("sentiment: " + str(sentence.labels))

# Nähdään, että yo. lauseen tunnesävy on hyvin suurella todennäköisyydellä 
# negatiivinen
```

```python
# Testataan tunnesävyanalyysia vielä muutamalla Airbnb-asuntojen arviolla

def sentence_sentiment(sentence):
  # poistetaan kommenteista mahdolliset rivinvaihtomerkit \n ja \r
  sentence = sentence.replace("\n", "")
  sentence = sentence.replace("\r", "")
  
  flair_sentence=flair.data.Sentence(sentence, use_tokenizer=True)
  sentiment_model.predict(flair_sentence)
  return flair_sentence.labels

# Luetaan Airbnb-käyttäjien arvostelut Wienin kohteista tammikuulta 2020
print("Luetaan Airbnb-data...")
review_data=pd.read_csv('http://data.insideairbnb.com/austria/vienna/vienna/2020-01-14/data/reviews.csv.gz')
print("Määritetään tunnesävyt kolmelle ensimmäiselle kommentille...")

# Poimitaan kolme ensimmäistä kommenttia ja tulostetaan ne
three_first=review_data[:3].copy()
pd.set_option('display.max_colwidth', None)
print(three_first.comments)
print("\nKommenttien tunnesävyt:\n==========================\n")

# Määritetään tunnesävyt kommenteille (tunnesävy ja sen todennäköisyys)
three_first['sentiment'] = three_first.apply(lambda row: sentence_sentiment(row.comments), axis=1)
pd.set_option('display.max_colwidth', 30)
print(three_first)
```

Nähdään, että yo. kolme kommenttia on luokiteltu aivan oikein tunnesävyltään positiivisiksi.

```python
import libvoikko
import string

voikko = libvoikko.Voikko(u"fi")
# esimerkkilause on Hesarin uutisotsikko toukokuun 11. päivältä:
lause = "Helleraja rikkoutunut, lämpöaalto hellii lähes koko maata ainakin helatorstaihin"
# muunnetaan lauseen kirjaimet pieniksi
lause = lause.lower()
# poistetaan pisteet, pilkut yms. välimerkit
lause = lause.translate(str.maketrans('', '', string.punctuation))
# poimitaan lauseen yksittäiset sanat listaan
sanat = lause.split(" ")

# tulostetaan sanojen perusmuodot
for sana in sanat:
  v=voikko.analyze(sana)
  if v:
    print(sana + " -> " + v[0]['BASEFORM'])
  else:
    print(sana + " -> VIRHE: perusmuotoa ei löydy!")

```

Ohjaamaton koneoppiminen poikkeaa ohjatusta oppimisesta siinä, että mallin opetusdatassa ei ole annettuna oikeita luokkia eli "oikeita vastauksia". Päämääränä ohjaamattomassa oppimisessa on löytää datasta jonkinlainen rakenne (esim. samankaltaisten tapausten ryppäitä eli klustereita). Koska oikeat luokat eivät ole tiedossa, mallin toimivuuden arviointi on huomattavasti vaikeampaa kuin ohjatun oppimisen tapauksessa. Yksi ohjaamattoman oppimisen menetelmä on ryvästäminen eli klusterointi, jossa lähtödata jaetaan mahdollisimman samankaltaisiin, mutta mahdollisimman etäällä toisistaan oleviin klustereihin. Klusterointi perustuu datapisteiden välisen etäisyyden mittaamiseen eli käytännössä numeerisiin piirteisiin. Ryvästämäisen lisäksi luennolla mainittuja ohjaamattoman oppimisen otteita ovat aihemallinnus, verkostoanalyysi sekä ostoskorianalyysi.

Ohjaamattoman oppimisen työnkulku etenee tiivisettynä seuraavasti, tässä esimerkissä metodina k-means klusterointi:
1. "Raakadata" siivotaan ja siitä jalostetaan piirteet ns. piirrematriisiksi.
2. Koneoppimismalli opetetaan piirrematriisin avulla -> klusterit (parametrina klustereiden määrä k)
3. Opetetun mallin suorituskykyä arvioidaan testidatan avulla
4. Malli viedään tuotantoon. Uudesta datasta irrotetaan piirrevektori, joka syötetään ennustemalliin. Ennustemallin ulostulona saadaan klusterin numero, johon sisääntulona annettu datapiste kuuluu.

Luennolla läpikäydyssä esimerkissä hyödynnettiin muuttujien normalisointia. Normalisoinnin ideana on skaalata data siten, että muuttujan arvot ovat nollan ja ykkösen välillä. Luennon esimerkissä hyödynnetty min-max scaler on herkkä datassa mahdollisesti esiintyville poikkeamille (outliers), minkä vuoksi poikkeamat tulisi poistaa datasta tai käytettää outliereille vähemmän herkkää normalisointimenetelmää, kuten <i>robust scaler</i>.

Ohjaamattoman oppimisen sovelluskohteita liiketoiminnassa ovat esimerkiksi asiakassegmentointi eli asiakkaiden jakaminen samankaltaisiin klustereihin sekä ostoskorianalyysi, jota voidaan hyödyntää esimerkiksi myymälän layoutin suunnittelussa. Verkostoanalyysia on hyödynnetty markkinoinnissa mm. niin sanottujen <a href="https://www.emerald.com/insight/content/doi/10.1108/EJMBE-10-2017-020/full/pdf?title=marketing-and-social-networks-a-criterion-for-detecting-opinion-leaders">mielipidevaikuttajien tunnistamiseen</a>. Aihemallinnusta on puolestaan sovellettu esimerkiksi asiakkaiden tukipyyntöjen <a href="https://monkeylearn.com/blog/introduction-to-topic-modeling/">kiireellisyyden luokittelussa</a>.

*Demo viikon 6 aikana oppimistani asioista*

Koska luennolla käsiteltiin klusterointia varsin kattavasti, halusin perehtyä tämän viikon demossa johonkin toiseen ohjaamattoman koneoppimisen lähestymistapaan. Lähdin perehtymään tarkemmin ostoskorianalyysin, jossa hyödynnetään <a href="https://towardsdatascience.com/apriori-association-rule-mining-explanation-and-python-implementation-290b42afdfc6">apriori-algoritmia</a>. Algoritmia käytetään assosiaatiosääntöjen louhintaan (<i>association rule mining</i>), jossa tarkoituksena löytää taustalla vaikuttavia yhteyksiä eri asioiden välillä. Klassinen esimerkki (vai urbaani legenda?) assosiaatiosääntöjen louhinnasta on löydös, jonka mukaan olutta ja vaippoja ostetaan usein yhdessä.


```python
# Ohjaamattoman oppimisen sovelluksena/demona tarkastellaan ostoskorianalyysia, jonka 
# toteutuksessa hyödynnetään apriori-algoritmia

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Luetaan datasetti Githubista (Groceries dataset)
# https://github.com/stedy/Machine-Learning-with-R-datasets/blob/master/groceries.csv
# Tiedoston yksi rivi sisältää kaupan yhden ostotapahtuman (transaktion) sisältämät tuotteet
# Tehdään ostoskorianalyysi datan pohjalta.

GROCERIES_DATA='https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/groceries.csv'
df=pd.read_csv(GROCERIES_DATA, header=None, names=range(32))

print(df)

# muunnetaan dataframe listaksi
basket_data=df.stack().groupby(level=0).apply(list).tolist()
print(basket_data)

# datasetissä on yhdeensä 9835 transaktiota
```

```python
# Muunnetaan kategorinen data "one-hot" enkoodattuun muotoon
te = TransactionEncoder()
te_ary = te.fit(basket_data).transform(basket_data)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Lasketaan datasta assiosiaatiosäännöt apriori-metodilla
# tässä apuna käytetty: 
# https://medium.com/swlh/a-tutorial-about-market-basket-analysis-in-python-predictive-hacks-497dc6e06b27

frequent_itemsets = apriori(df, min_support = 0.002, max_len = 3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules['lhs items'] = rules['antecedents'].apply(lambda x:len(x) )

# Säännöt, joissa sekä lift että confidence ovat suuria, ovat todennäköisimpiä
rules[(rules['lift'] >= 5) & (rules['confidence'] >= 0.5)].sort_values(by='lift', ascending=False)
```

Yllä louhittujen sääntöjen perusteella voidaan mm. havaita, että jos ostoskorissa on voita ja juustoa, sinne todennäköisesti päätyy myös  kermavaahtoa/hapankermaa. Vastaavasti juureksia ostetaan usein yhdessä yrttien ja jogurtin kanssa. Ostoskorianalyysin tietoja kauppias voi hyödyntää esimerkiksi miettiessään tuotteiden sijoittelua myymälässä.

Datatieteen kontekstissa visualisoitava tieto on tyypillisesti dataa tai informaatiota. Visualisoinnin avulla "raakadata" voidaan jalostaa informaatioksi. Havainnollistamisen ja valitun esitystavan kautta informaatio voi jalostua ihmisen toiminnan seurauksena korkeamman tason tiedoksi eli ymmärrykseksi ja tietämykseksi.

Visualisoinnin suunnittelija muuntaa lähtötiedon suunnitteluprosessissa informaatioksi, joka välitetään käyttäjälle visuaalisia vihjeitä ja käyttöliittymän toimintoja hyödyntäen. Käyttäjä muodostaa eri näkymien ja vuorovaikutteisten toimintojen perusteella informaatiosta oman tulkintansa eli tietämystä. Kyseessä on siis tietojohtamisen SECI-mallin mukainen sisäistämisvaihe (internalization), jossa eksplisiittinen tieto muuntuu ymmärryksen kautta yksilön hiljaiseksi tiedoksi ja tulee osaksi hänen henkilökohtaista tietopohjaansa.

Visualisointia hyödynnetään yleisesti keinona, jolla informaatiota (mm. analyysin tulokset) kommunikoidaan esimerkiksi päätöksentekijöille tai muille loppukäyttäjille. Datatieteessä visualisointia hyödynnetään myös "raakadatan" tutkivassa kartoittamisessa eli eksploraatiossa. Visuaalisessa muodossa aineistosta voi paljastua piirteitä, jotka voisivat jäädä helposti piiloon teksti- tai taulukkomuodossa. Koponen et al. (2016) määrittelevätkin visualisoinnin eksploratiiviseksi eli uusia piirteitä aineistosta paljastavaksi grafiikaksi. Näin ollen visualisoinnin tehtävä painottuu uuden tiedon löytämiseen, eikä pelkästään tiedon välitykseen.

Luennolla käytiin läpi <a href="https://www.amazon.com/Visualizing-Data-Explaining-Processing-Environment/dp/0596514557">Ben Fry:n</a> (2008) datan visualisoinnin seitsemän askeleen prosessi, joka on analoginen datatieteen työprosessin kanssa. Prosessi alkaa datan keruulla, jonka jälkeen dataa esikäsitellään, mallinnetaan piirteisiin ja luokkiin sekä tarvittaessa suodatetaan. Seuraavassa vaiheessa hyödynnetään tiedon louhinnan menetelmiä, kuten klusterointia, jonka avulla voidaan tunnistaa aineiston rakenteita. Seuraavaksi valitaan tilanteeseen sopiva visualisointitapa, jonka jälkeen perusvisualisointia jalostetaan "houkuttelevammaksi" ja rakennetaan siihen vuorovaikutteisuutta.

Jotta visualisointi onnistuisi tavoitteessaan eli informaation välittämisessä ja ymmärryksen lisäämisessä, tulee visualisointien suunnittelussa huomioida tiettyjä asioita. Anne Tervakarin <a href="https://www.slideshare.net/AnneTervakari/visualisointien-suunnittelua">esityksessä</a> on kuvattu visualisointiprosessia sekä visualisointien suunnittelussa huomioitavia asioita. Edellä kuvattu Fry (2008):n prosessi keskittyy visualisointien tekniseen toteutukseen, kun taas Tervakarin esittämä prosessimalli mukailee enemmän CRISP-DM:ää. Tervakarin kuvaamassa mallissa lähdetään liikkeelle liiketoiminnan ongelman tai tavoitteen määrittelystä. Alkuvaiheessa määritellään lisäksi toimintaympäristö sekä loppukäyttäjät ja heidän tarpeensa. Tervakarin esityksessä on havainnollistettu myös erilaisiin tapauksiin soveltuvan esitystavan valintaa. Esitystavan valinnassa on syytä muistaa välttää visualisoinnin tyypillisimpiä <a href="https://www.kdnuggets.com/2017/10/5-common-mistakes-bad-data-visualization.html">sudenkuoppia</a>.

Kun visuaalinen viestintä siirtyy yhä enenevässä määrin paperilta tietokoneen tai älypuhelimen ruudulle, on vuorovaikutteisuudella yhä merkittävämpi rooli visualisointien käyttökokemuksessa. Esimerkiksi <a href="https://medium.com/design-ibm/designing-for-data-visualization-c2b18359878c">IBM:llä</a> visualisointien suunnittelun lähtökohtana on staattisten kuvien sijaan interaktiivinen kokemus, joka antaa käyttäjälle mahdollisuuden porautua syvemmälle dataan ja tehdä eksploratiivista analyysia itsenäisesti. Vuorovaikutteisuus tukee siten informaation tutkimista ja ymmärtämistä.

Yksi visualisoinnin keino, jota luennolla ei varsinaisesti käsitelty on datapohjainen tarinankerronta (<a href="https://www.youtube.com/watch?v=r5_34YnCmMY">Storytelling with data</a>). Tarinankerronnassa visualisointi rakennetaan tarinan muotoon, jossa on tietty alkutilanne (<i>setup</i>), keskikohta (<i>conflict</i>) ja loppu (<i>resolution</i>). Datapohjaisessa tarinankerronnassa tavoitteena on tukea informaation sisäistämistä toivotulla tavalla hyödyntämällä sekä kuva- että tekstimuotoista informaatiota. Kuvalliset elementit auttavat hahmottamaan aineiston rakennetta ja muodostamaan ilmiöstä nopean kokonaiskuvan. Tekstimuotoisella informaatiolla pystytään taas välittämään tarkkaa tai abstraktia analyyttista tietoa. Tekstin avulla voidaan viestiä visualisoinnin olennaisin pointti ja opastaa syvemmälle menevään tulkintaan.

Knaflic (2015) kuvaa tarinankerronnan rakentamisen prosessin seuraavasti: 
1. Ymmärrä konteksti. Kuka on yleisösi ja millaista tietoa he tarvitsevat?
2. Valitse tilanteeseen sopiva visuaalinen kuvaustapa. 
3. Eliminoi sotku (<i>clutter</i>). Poista visualisoinnista kaikki ylimääräinen/turha informaatio.
4. Suuntaa yleisön huomio olennaisimpaan. Tässä voi esimerkiksi hyödyntää värejä havainnollisella tavalla (esim. vähemmän tärkeä informaatio harmaalla värillä ja tärkeä oranssilla).
5. Ajattele kuin suunnittelija (<i>Think like a designer</i>). Mitä haluat, että loppukäyttäjä pystyy tekemään datan perusteella? Millainen visualisointi mahdollistaa tämän?
6. Kerro tarina. Rakenna tarina datan ympärille (alku-keskikohta-loppu).

- Kaikkein tärkein sääntö visualisointeja laatiessa on: <i>Valitse esitystapa, joka tuottaa mahdollisimman selkeän lopputuloksen</i> (Koponen et al. 2016). Visualisoinnin tulee esittää selkeimmällä mahdollisella tavalla oikeita ja luotettavasta lähteestä kerättyjä tietoja. Sen tulee sisältää tietoa mahdollisimman paljon, muttei yhtään liikaa. Visualisoinnin tulee kiinnittää lukijan huomio kaikkein merkittävimpiin seikkoihin. Visualisoinnin esitystapa tulee valita siten, ettei se johda lukijaa harhaan ja vie huomiota aineiston todellisilta piirteiltä.
- Vältä yleisimpiä visualisoinnin <a href="https://www.kdnuggets.com/2017/10/5-common-mistakes-bad-data-visualization.html">sudenkuoppia</a>. 1.) Jos lähtödata on huonoa, visualisoinnilla se ei paremmaksi muutu. 2.) Valitse tilanteeseen sopiva visualisointi. Esim. piirakkakaavio ei ole hyvä valinta, jos tarkoitus on havainnollistaa suhteellisia osuuksia. 3.) Ei liikaa informaatiota eikä "värien ilotulitusta". 4.) Älä hämää loppukäyttäjää katkaisemalla y-akselia. Y-akseli alkaa nollasta (tai negatiivisesta arvosta).  5.) Käytä x- ja y-akseleilla johdonmukaista asteikkoa.

Demoillaan interaktiivisen visualisoinnin rakentamista Pythonin PyPlot-kirjastolla.

```python
# Demo 7: Visuaalista analytiikkaa/analyysia Pythonin PyPlot-kirjastoa hyödyntäen

# Visualisoidaan Wienin Airbnb-asuntodataa kaupunginosakohtaisesti

import plotly.express as px
import pandas as pd

# luetaan Wienin Airbnb-data tammikuulta 2020
df=pd.read_csv('http://data.insideairbnb.com/austria/vienna/vienna/2020-01-14/data/listings.csv.gz')
df.describe()
```

```python
# poistetaan dollarimerkki hintasarakkeesta (price)
df['price'] = df['price'].replace('[\$,)]','',  \
  regex=True).replace('[(]','-', regex=True).astype(float)

# Lasketaan kunkin kaupunginosan Airbnb-asuntojen määrä sekä hinnan keskiarvo ja mediaani 
df_price=df[['neighbourhood_cleansed', 'price']].groupby(['neighbourhood_cleansed']).agg(['count', \
                                                                                          'mean', \
                                                                                          'median'])

df_price.columns=df_price.columns.droplevel()
df_price=df_price.sort_values(ascending=True, by='count')
print(df_price)

# Piirretään horisontaalinen pylväsdiagrammi asuntojen määrästä eri kaupunginosissa
# interaktiivisuuden määrä tässä kuvaajassa on varsin minimaalinen; viemällä hiiren
# kursorin kaupunginosan pylvään päälle, käyttäjä saa lisätietoa alueen
# Airbnb-asuntojen keski- ja mediaanihinnoista.

fig = px.bar(df_price,  x='count', y=df_price.index, orientation='h', title='Airbnb asuntojen määrät Wienissä kaupunginosittain',
             labels={'count':'asuntojen määrä', 'y':'kaupunginosa', 'median':'asunnon mediaanihinta (eur)', 'mean':'asunnon keskiarvohinta (eur)'}, 
             hover_data=['count','median','mean'])
fig.show()
```

```python
# Lopuksi demoillaan vielä Plotly Expressin interaktiivista visualisointia 
# aikasarjadatalla.
# Koodi on poimittu osoitteesta https://plotly.com/python/sliders/
# Esimerkkikoodi hyödyntää Gapminder-datasettiä.

df2 = px.data.gapminder()
fig2 = px.scatter(df2, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])

fig2["layout"].pop("updatemenus") # optional, drop animation buttons
fig2.show()
```
